# Gini Index Formula
## Gini(p) = 1 - Σ p(i)²


# Imports

```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn import tree
from sklearn import metrics
from sklearn.datasets import load_iris
```

# Load Iris Data

```python
data =load_iris()
print(data)
```

```python
df = pd.DataFrame(data.data,columns=data.feature_names)
df['target']=data.target
```

# Training data and Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(df[data.feature_names],df['target'],random_state=0,test_size=0.20)
```

# Decision Tree Classifier

```python
clf = DecisionTreeClassifier(max_depth=3,criterion='gini')
clf.fit(X_train,Y_train)
```

```python
fn = ['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn = ['setosa','versicolor','virginica']
```

# Plotting Tree

```python
fig,axes = plt.subplots(nrows=1,ncols=1,figsize=(4,4),dpi=300)
tree.plot_tree(clf,feature_names=fn,class_names=cn,filled=True)
fig.savefig('mytree.png')

```

# Prediction

```python
print("-------------------------------")
prediction = clf.predict(X_test)
print(prediction)
print("-------------------------------")
```

# Accuracy

```python
print("Accuracy ",metrics.accuracy_score(Y_test,prediction))
print(metrics.classification_report(Y_test,prediction))
print(metrics.confusion_matrix(Y_test,prediction))
```

# Conclusion

## In this Practical we make decision tree using decision tree classifier with gini index algorithm. We can show that taking depth upto 3 with gini index criteria we achieve a very good accuracy. Gini index favours large partitions, A feature with lowest gini index choose for split.
